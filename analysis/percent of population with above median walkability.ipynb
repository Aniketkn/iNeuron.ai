```python
import pandas as pd
import os

hex_data = os.path.abspath('../process/data/output/global_indicators_hex_250m_2021-06-21.csv')
df = pd.read_csv(hex_data)
df
```

```python
df.all_cities_walkability.hist()
```

```python
df.describe()
```

```python
df['walkability_above_average'] = (df['all_cities_walkability'] > df.describe().loc['mean','all_cities_walkability']).astype(int)
df['walkability_above_median'] =  (df['all_cities_walkability'] > df.describe().loc['50%' ,'all_cities_walkability']).astype(int)
```

```python
for stat in ['walkability_above_average','walkability_above_median']:
    print(f"\n{stat}:")
    df_wa = df.groupby(['Continent','Country','City',stat])['pop_est'].sum().unstack(level=3)
    df_wa = df_wa.rename(columns = {0:f'not_{stat}',1:f'{stat}'})
    df_wa['Total'] = df_wa.sum(axis=1)
    df_wa[f"pct_{stat}"] = 100*(df_wa[f'{stat}']/df_wa['Total'])
    print(df_wa[f"pct_{stat}"])
    df_wa.to_csv(f"Global Indicators - 2021-06-21 - percentage of population - {stat}")

```
