## Create study region polygon
This notebook shows the process to load administrative boundary (assume it is pre-exist), and create a 10km buffer around the boundary to set a study region 

```python
# Libraries used for OSMnx analyses and output
import matplotlib.pyplot as plt
import pandas as pd
import geopandas as gpd
from shapely.geometry import shape,Point, LineString, Polygon
```

### Set up project config

```python
places = {'phoenix' : 'Phoenix, Arizona, USA'} # study region name

time = '_201905' # output data time

buffer_dist = 1e4 #study region buffer 10km

data_folder = '../data/Maricopa_County'

OSM_folder = '../data/OSM'

shape_filename = 'MC_Boundary.shp'

filepath = data_folder + '/' + shape_filename
```

## Get place boundary

```python
def get_bufferedstudyregion_polygon(filepath, buffer_dist=1e4):
    shape = gpd.GeoDataFrame.from_file(filepath)
    polygon = shape['geometry'].iloc[0]
    buffer_polygon = polygon.buffer(buffer_dist)
    return buffer_polygon
```
